1. El promedio general de los salarios, desviaci√≥n est√°ndar, quintiles y rango. (hint:
escoge la columna adecuada para comparar)

```python
# Primero importamos la librer√≠a de pandas.
import pandas as pd

# Creamos el dataframe usando 'pd.read_csv()'.
df= pd.read_csv('ds_salaries.csv',delimiter=',')

# Mostramos el dataframe creado.
df
```

```python
# Seleccionamos la columna para comparar (salary_in_usd).
salarios = df["salary_in_usd"]

# Raz√≥n por la que elegimos esa columna:
# Si analiz√°ramos la columna salary, tendr√≠amos sueldos en diferentes monedas, lo que har√≠a que los valores no fueran comparables entre s√≠.

# Realizamos los c√°lculos estad√≠sticos solicitados.
promedio = salarios.mean()
desviacion = salarios.std()
quintiles = salarios.quantile([0, 0.2, 0.4, 0.6, 0.8, 1])
rango = salarios.max() - salarios.min()

# Mostramos los resultados.
print("üìä An√°lisis general de salarios en USD")
print(f"Promedio: {promedio:,.1f}")
print(f"Desviaci√≥n est√°ndar: {desviacion:,.1f}")
print(f"Rango: {rango:,.1f}")
print("\nQuintiles:")
print(quintiles)
```

2. Una comparaci√≥n entre los salarios agrupados por 3 diferentes categor√≠as a tu
elecci√≥n. ¬øPara cu√°l de ellas las medidas de tendencia central son m√°s
representativas? ¬øPara cu√°l son menos representativas? Explica.

```python
# Creamos una lista con las categor√≠as que elegimos para comparar.
categorias = ["experience_level", "company_size", "remote_ratio"]

# Creamos un diccionario vac√≠o para guardar los resultados.
resultados = {}

# Creamos un ciclo para que recorra la lista con las categor√≠as elegidas.
for cat in categorias:
    resumen = (
        df.groupby(cat)["salary_in_usd"] # Agrupamos por categor√≠as.
        .agg(["mean", "median", "std", "count"]) # Indicadores que elegimos para realizar la comparaci√≥n.
        .sort_values("mean", ascending=False) 
    )
    resultados[cat] = resumen
    
# Mostramos los resultados.
    print(f"\nüìä An√°lisis por categor√≠a: {cat}")
    print(resumen)
```

####  Explicaci√≥n de resultados
##### 1) Comparaci√≥n por nivel de experiencia (experience_level):
Los niveles de experiencia suelen mostrar diferencias salariales muy marcadas.
En general, el an√°lisis nos muestra que:
Los perfiles EX (Executive) y SE (Senior) presentan los salarios promedio m√°s altos.
Los perfiles EN (Entry Level) tienen salarios claramente m√°s bajos.
La mediana suele es representativa porque los grupos no presentan tanta dispersi√≥n extrema.
La desviaci√≥n est√°ndar es moderada: dentro de cada nivel de experiencia, las variaciones salariales son relativamente consistentes.
###### Conclusi√≥n: Es una categor√≠a donde la media y la mediana describen muy bien los datos.
##### 2) Comparaci√≥n por tama√±o de empresa (company_size):
En esta categor√≠a aparecen tres grupos: S (small), M (medium), L (large).
Los resultados suelen mostrar:
Las empresas grandes (L) pagan los salarios m√°s altos.
Las empresas peque√±as (S) pagan menos.
Las empresas medianas (M) se ubican en un punto intermedio.
Pero aqu√≠ la desviaci√≥n est√°ndar es alta, sobre todo en empresas grandes, donde conviven sueldos muy altos (cargos ejecutivos) y sueldos bajos (puestos junior).
Entonces ¬øSon representativas las medidas de tendencia central? Las medidas siguen siendo √∫tiles, pero ya no son tan precisas.
Esto se debe a:
Alta variabilidad dentro de cada grupo, especialmente en empresas grandes; Mezcla de muchos tipos de puestos dentro de una misma categor√≠a.
###### Conclusi√≥n: Las estad√≠sticas resumen bien la tendencia, pero la dispersi√≥n indica que no todos los sueldos son comparables dentro del grupo.
##### 3) Comparaci√≥n seg√∫n modalidad de trabajo remoto (remote_ratio):
Grupos:
0 ‚Üí presencial.
50 ‚Üí h√≠brido.
100 ‚Üí remoto total.
Aqu√≠ suele verse:
Los salarios remotos al 100% tienden a ser m√°s altos (especialmente si la empresa es de EE.UU. y contrata talento extranjero).
El grupo 0 (presencial) tiene m√°s dispersi√≥n y suele incluir sueldos inferiores.
El grupo 50 (h√≠brido) queda en un punto medio.
Sin embargo, este es el grupo donde las medidas de tendencia central son menos representativas.¬øPor qu√©?
Porque:
El trabajo remoto est√° muy asociado a industrias globales y empresas que pagan en d√≥lares.
El presencial puede incluir pa√≠ses con salarios mucho m√°s bajos.
Hay mezclas muy fuertes entre pa√≠ses, monedas y roles.
Esto produce alt√≠sima variabilidad en salarios dentro del mismo grupo.
###### Conclusi√≥n: Las medias y medianas no describen tan bien el grupo porque los datos son demasiado heterog√©neos.

3. ¬øQu√© cargos reciben mejores sueldos en las empresas con sede en Estados Unidos?

```python
# Filtramos empresas cuya sede est√° en Estados Unidos (US).
usa = df[df["company_location"] == "US"]

# Ordenanos por salario en USD de mayor a menor y seleccionamos los primeros 10.
top10 = usa.sort_values("salary_in_usd", ascending=False).head(10)

# Mostramos los resultados.
print("Los 10 salarios m√°s altos en empresas con sede en EE.UU.:")
top10[["job_title", "salary_in_usd", "experience_level", "company_size", "remote_ratio"]]
```
